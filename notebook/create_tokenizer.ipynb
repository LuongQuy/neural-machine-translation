```
processed_data_path = 'processed_text_data/'
processed_en_train_path = processed_data_path + 'train/train.en'
processed_vi_train_path = processed_data_path + 'train/train.vi'

processed_en_dev_path = processed_data_path + 'dev/tst2012.en'
processed_vi_dev_path = processed_data_path + 'dev/tst2012.vi'

processed_en_test_path = processed_data_path + 'test/tst2013.en'
processed_vi_test_path = processed_data_path + 'test/tst2013.vi'

processed_en_vocab_path = processed_data_path + 'vocab/vocab.en'
processed_vi_vocab_path = processed_data_path + 'vocab/vocab.vi'
```

```
import tensorflow as tf
import pickle
```

```
def load_data(path):
  try:
    with open(path, 'r') as f:
      data = f.read().split('\n')
  except NameError:
    print('Error:', NameError)
  return data
```

```
en_train = load_data(processed_en_train_path)
en_dev = load_data(processed_en_dev_path)
en_test = load_data(processed_en_test_path)
en_vocab = load_data(processed_en_vocab_path)
```

```
vi_train = load_data(processed_vi_train_path)
vi_dev = load_data(processed_vi_dev_path)
vi_test = load_data(processed_vi_test_path)
vi_vocab = load_data(processed_vi_vocab_path)
```

```
en_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='')
en_tokenizer.fit_on_texts(en_train)
en_tokenizer.fit_on_texts(en_dev)
en_tokenizer.fit_on_texts(en_test)
en_tokenizer.fit_on_texts(en_vocab)
print('size of vocab:', len(en_tokenizer.word_index))
```

```
vi_tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='')
vi_tokenizer.fit_on_texts(vi_train)
vi_tokenizer.fit_on_texts(vi_dev)
vi_tokenizer.fit_on_texts(vi_test)
vi_tokenizer.fit_on_texts(vi_vocab)
print('size of vocab:', len(vi_tokenizer.word_index))
```

```
# save tokenizer to file
with open('tokenizer/tokenizer.pickle', 'wb') as handle:
    pickle.dump(
        {
         'en_tokenizer': en_tokenizer,
         'vi_tokenizer': vi_tokenizer,
        }, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
