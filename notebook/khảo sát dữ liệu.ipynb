```
import matplotlib.pyplot as plt
```

```
en_train_data_path = 'text_data/train/train.en'
vi_train_data_path = 'text_data/train/train.vi'
```

```
# Khảo sát độ dài câu
def plot_hist(arr, title):
  plt.hist([len(s.split(" ")) for s in arr], bins=100)
  plt.title(title)
  plt.xlabel('Len (number of token)')
  plt.ylabel('# samples')
  plt.show()

with open(en_train_data_path, 'r', encoding='UTF-8') as f:
    en_data = f.readlines()
    plot_hist(en_data, 'Histogram độ dài câu tiếng anh')
with open(vi_train_data_path, 'r', encoding='UTF-8') as f:
    vi_data = f.readlines()
    plot_hist(vi_data, 'Histogram độ dài câu tiếng việt')
```
