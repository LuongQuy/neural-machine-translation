```
from underthesea import word_tokenize
import unicodedata
import re

def preprocess_sentence(s):
  s = word_tokenize(s, format="text")
  s = s.lower()
  s = re.sub(r'[" "]+', " ", s)
  s = s.strip()
  return s

def create_new_data(path, new_path, num_examples):
    with open(path, 'r', encoding='UTF-8') as f:
        data = f.readlines()
        data = data[:num_examples]
        data = [preprocess_sentence(sent) for sent in data]
        data = '\n'.join(data)
        with open(new_path, 'w', encoding='UTF-8') as f2:
            f2.write(data)
    print(path + '--> Done')

num_examples = None

# create_new_data(en_train_data_path, processed_en_train_path, num_examples)
create_new_data(vi_train_data_path, processed_vi_train_path, num_examples)

# create_new_data(en_dev_data_path, processed_en_dev_path, num_examples)
create_new_data(vi_dev_data_path, processed_vi_dev_path, num_examples)

# create_new_data(en_test_data_path, processed_en_test_path, num_examples)
create_new_data(vi_test_data_path, processed_vi_test_path, num_examples)

# create_new_data(en_vocab_path, processed_en_vocab_path, num_examples)
create_new_data(vi_vocab_path, processed_vi_vocab_path, num_examples)
```
